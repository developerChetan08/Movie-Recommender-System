```python
## Real Format:- '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
## Change Correct Format:- ['Action','Adventure','Fantasy','Science Fiction']
```

```python
##First String of list to convert Normal List:-
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# 3)Cast Col:- same process but get only 3 actors
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L    
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
# 4)Crew Col:- get only job = director name
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#ps.stem('loving')
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
def recommend(movie):
    ##fetch indexing
    movie_index = new_df[new_df['title'] == movie].index[0]
    ##calculate distances
    distances = similarity[movie_index]
    ##sort distances array and fetch only 5 records
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    ##then print movie list
    for i in movies_list:
        print(new_df.iloc[i[0]].title)   
    
```

```python
recommend('Batman Begins')
```
